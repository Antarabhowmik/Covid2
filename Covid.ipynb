#### 6.1 Redefine Features, Target, and Split Data for Tuning

```python
from sklearn.model_selection import train_test_split

# Redefine features (X) and target (y) from the original DataFrame `df`
# This ensures the split is consistent with the pipeline's expected raw input
X_raw = df.drop('mortality', axis=1)
y_raw = df['mortality']

# Split data into training and testing sets
# Use stratify=y_raw to maintain the class distribution in both sets due to imbalance
X_train_raw, X_test_raw, y_train_raw, y_test_raw = train_test_split(X_raw, y_raw, test_size=0.2, random_state=42, stratify=y_raw)

print("X_train_raw shape:", X_train_raw.shape)
print("X_test_raw shape:", X_test_raw.shape)
print("y_train_raw shape:", y_train_raw.shape)
print("y_test_raw shape:", y_test_raw.shape)

print("\nDistribution of mortality in y_train_raw:")
display(y_train_raw.value_counts(normalize=True))
print("\nDistribution of mortality in y_test_raw:")
display(y_test_raw.value_counts(normalize=True))
```

#### 6.2 Build Imbalanced-Learn Pipeline

```python
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE

# Identify numerical and categorical features (from X_raw)
numerical_features_raw = X_raw.select_dtypes(include=['int64', 'float64']).columns
categorical_features_raw = X_raw.select_dtypes(include=['object']).columns

# Create preprocessing pipelines for numerical and categorical features
numerical_transformer_raw = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_transformer_raw = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Create a column transformer to apply different transformations to different columns
preprocessor_raw = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer_raw, numerical_features_raw),
        ('cat', categorical_transformer_raw, categorical_features_raw)
    ])

# Create the imbalanced-learn pipeline with SMOTE and RandomForestClassifier
pipeline_with_smote = ImbPipeline(steps=[
    ('preprocessor', preprocessor_raw),
    ('smote', SMOTE(random_state=42)),
    ('classifier', RandomForestClassifier(random_state=42))
])

print("Imbalanced-Learn Pipeline created successfully.")
```

#### 6.3 Hyperparameter Tuning with GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, f1_score, recall_score

# Define parameter grid for GridSearchCV
param_grid = {
    'smote__sampling_strategy': [0.5, 0.75, 1.0], # Ratio of minority to majority samples after resampling
    'classifier__n_estimators': [100, 200, 300], # Number of trees in the forest
    'classifier__max_depth': [10, 20, None], # Maximum depth of the tree
    'classifier__min_samples_split': [2, 5], # Minimum number of samples required to split an internal node
    'classifier__min_samples_leaf': [1, 2] # Minimum number of samples required to be at a leaf node
}

# Use f1_score as the scoring metric for GridSearchCV, as it balances precision and recall
# Make sure to specify pos_label=1 if your minority class is 1
scorer = make_scorer(f1_score, pos_label=1)

# Initialize GridSearchCV
grid_search = GridSearchCV(
    estimator=pipeline_with_smote,
    param_grid=param_grid,
    scoring=scorer,
    cv=5, # 5-fold cross-validation
    n_jobs=-1, # Use all available cores
    verbose=2
)

print("Starting GridSearchCV...")
# Fit GridSearchCV on the training data
grid_search.fit(X_train_raw, y_train_raw)

print("GridSearchCV complete.")
print(f"Best parameters found: {grid_search.best_params_}")
print(f"Best F1-Score: {grid_search.best_score_:.4f}")
```

### 7. Final Production-Ready ML Pipeline (with Tuned Parameters)

```python
# Import necessary libraries
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
import joblib
import pandas as pd

# Separate features (X) and target (y) from the original DataFrame `df`
X = df.drop('mortality', axis=1)
y = df['mortality']

# Identify numerical and categorical features
numerical_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X.select_dtypes(include=['object']).columns

# Create preprocessing pipelines for numerical and categorical features
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Create a column transformer to apply different transformations to different columns
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Define the final model using the BEST PARAMETERS found from GridSearchCV
# We remove SMOTE from the final production pipeline as it's only needed for training and tuning
# The final model is trained on the full dataset with balanced class weights if tuning suggested it
# or with other parameters if specified.
# For this production pipeline, we will use the best classifier parameters found and apply `class_weight='balanced'`
# directly to the classifier if it was not part of the grid search or if it generally performs better in production.
# However, if SMOTE was crucial, it might be better to retrain a pipeline with SMOTE as part of it here on full data,
# or simply use the best_pipeline from GridSearchCV directly if it was trained on all data.

# Since GridSearchCV involved SMOTE and RandomForest, the best parameters are for both.
# For the final production pipeline (trained on full data without explicit SMOTE here),
# we extract the best parameters for the RandomForestClassifier.

best_rf_params = grid_search.best_params_.copy()
# Remove SMOTE parameters if they exist in best_params_
for key in list(best_rf_params.keys()):
    if key.startswith('smote__'):
        del best_rf_params[key]
    else:
        # Rename classifier__ parameters to direct parameters for RandomForestClassifier
        best_rf_params[key.replace('classifier__', '')] = best_rf_params.pop(key)

# Add class_weight='balanced' explicitly as it's a good practice for imbalanced datasets
# if not explicitly tuned or if it was found beneficial during initial runs.
# This ensures the model learns from the minority class during the final training on full data.
if 'class_weight' not in best_rf_params: # Ensure we don't overwrite if it was tuned
    best_rf_params['class_weight'] = 'balanced'

classifier_tuned = RandomForestClassifier(random_state=42, **best_rf_params)

# Create the full pipeline with tuned classifier
production_pipeline_tuned = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', classifier_tuned)
])

# Train the pipeline on the full dataset
production_pipeline_tuned.fit(X, y)

print("Tuned Production Pipeline training complete.")
```

```python
# Import necessary libraries
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
import joblib
import pandas as pd

# Separate features (X) and target (y) from the original DataFrame `df`
X = df.drop('mortality', axis=1)
y = df['mortality']

# Identify numerical and categorical features
numerical_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X.select_dtypes(include=['object']).columns

# Create preprocessing pipelines for numerical and categorical features
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Create a column transformer to apply different transformations to different columns
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Define the final model (Random Forest Classifier with class weighting for imbalance)
classifier = RandomForestClassifier(random_state=42, class_weight='balanced')

# Create the full pipeline
production_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', classifier)
])

# Train the pipeline on the full dataset
production_pipeline.fit(X, y)

print("Pipeline training complete.")
```
